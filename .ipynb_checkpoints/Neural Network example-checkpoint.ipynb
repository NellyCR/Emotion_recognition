# Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.constraints import maxnorm
from keras.optimizers import Nadam
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

# Reading data

```python
# read just total_lines observations
total_lines = 1000
data=pd.read_csv('1Feb_9Jun-data.csv')[:1000]
data.drop(['timestamp', 'close_time', 'quote_av', 'tb_base_av', 'tb_quote_av', 'ignore'], axis=1, inplace=True)
data
```

# Preprocessing data

__Esta parte del código sólo córrela, no es importante saber qué pasa adentro porque sólo es el procesamiento de datos para este modelo en particular. Puede tardar un poco.__

Al final obtenemos los dataframes `X_processed` y `y_processed` para entrenar el modelo.

```python
from tqdm import tqdm

N = 60

X = np.zeros((len(data)-60,N*len(data.columns)))

for i in tqdm(range(len(data)-60)):
    new_line = np.array([])
    for col in data.columns:
        new_line = np.hstack((new_line, data[col].loc[i:i+59].values))
    X[i] = new_line

from statsmodels.tsa.ar_model import AutoReg


values_period = data['close'].loc[N:N+59].values
model = AutoReg(values_period,1).fit()
y = np.zeros((len(data)-60, 3))

for i in tqdm(range(N, len(data))):
    
    values_period = data['close'].loc[i:i+59].values
    
    try:
        ar_model = AutoReg(values_period,1).fit()
    except:
        print(i)
    
    alpha = ar_model.params[0]
    beta = ar_model.params[1]
    eps = ar_model.sigma2
    
    mu = alpha/(1-beta)
    a = -np.log(beta)
    sigma= ar_model.sigma2*np.sqrt((2*a)/(1-beta**2))
    
    y[i-N] = np.array([mu, a, sigma])

y_raw = pd.DataFrame(y, columns=['mu', 'a', 'sigma'])
y_processed = y_raw.loc[np.abs(y_raw['a'])>0.01].copy()

columns = []
for col in data.columns:
    for i in range(1,N+1):
        columns.append(col+str(i))

X_processed = pd.DataFrame(X, columns=columns)
X_processed = X_processed.loc[y_processed.index].copy()
```

# Normalizing data

```python
X_train, X_test, y_train, y_test = train_test_split(X_processed.values, 
                                                    y_processed.values, 
                                                    test_size=0.2, 
                                                    random_state=None)

scaler = StandardScaler()
X_train_normalized = scaler.fit_transform(X_train)
```

# Training neural network

```python
model = Sequential()
model.add(Dense(20, input_dim=N*len(data.columns), activation = 'relu'))
model.add(Dense(20, activation = 'relu'))
model.add(Dense(20, activation = 'relu'))
model.add(Dense(20, activation = 'relu'))
model.add(Dense(20, activation = 'relu'))
model.add(Dense(3))
model.compile(optimizer='adam', loss='mse', metrics=['mape'])

batch_size = 2**5
epochs = 100

# reducir el learning rate a la mitad (factor = 0.5)
reduceLR = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10, min_lr=9e-10, verbose=1)
earlyStopping = EarlyStopping(monitor='val_loss', patience=10)

callbacks = [reduceLR, earlyStopping]

history = model.fit(X_train, 
                    y_train, 
                    batch_size=batch_size, 
                    epochs=epochs, 
                    validation_split=0.1, 
                    callbacks=callbacks)
```

```python
plt.plot(history.history['mape'][10:])
plt.plot(history.history['val_mape'][10:])
```
